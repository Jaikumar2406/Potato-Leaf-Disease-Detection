```python
## cache , shuffel , prefatch
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
##resizing and rescaling
```

```python
resize_rescaling = tf.keras.Sequential([
    keras.layers.experimental.preprocessing.Resizing(120,120),
    keras.layers.experimental.preprocessing.Rescaling(1./255)
])
```

```python
##data
```

```python
data_augmanetation = keras.Sequential([
    keras.layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
    keras.layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
train_ds = train_ds.map(lambda x , y: (data_augmanetation(x , training = True) , y)).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
train_ds
```

```python
model = keras.Sequential([
    resize_rescaling,
    keras.layers.Conv2D(32 , kernel_size=(3 ,3) , activation='relu' ,input_shape =(16 , 120,120,3)),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(64 , kernel_size=(3 ,3) , activation='relu'),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(64 , kernel_size=(3 ,3) , activation='relu' ),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(64 , kernel_size=(3 ,3) , activation='relu'),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(64 , kernel_size=(3 ,3) , activation='relu' ),
    keras.layers.MaxPool2D((2,2)),

    keras.layers.Flatten(),
    keras.layers.Dense(64 , activation='relu' ),
    keras.layers.Dense(3 , activation='softmax')
])

```

```python
model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy'])
```

```python
history = model.fit(
    train_ds,
    batch_size=16,
    validation_data=val_ds,
    verbose=1,
    epochs=20
)
```

```python
model.save('model.h5')
```
